```python
import time
max_value = 100


pb2 = widgets.FloatProgress(min=0,
                        max=max_value,
                        value = 50,
                        description = 'Ucitavam:')
pb2.bar_style = 'success'

display(pb2)

def update_progress(value):
    pb2.value = value  # Ažuriramo vrijednost trake napretka

# Kreiranje slidera za interaktivnu promjenu vrijednosti
slider = widgets.FloatSlider(min=0, max=max_value, value=50, step=1, description='Vrijednost:')
slider.observe(lambda change: update_progress(change['new']), names='value')

# Prikazivanje slidera
display(slider)
while pb2.value < max_value:
    pb2.value += 0.1 # povecaj vrijednost trake
    time.sleep(0.01)
```
